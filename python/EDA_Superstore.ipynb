Data Analysis 

DATA QUALITY CHECK

```python
print("\n === DATA QUALITY CHECK === \n")

# check for missing values 
print("Missing Values: \n")
missing_data = df.isnull().sum()
print(missing_data[missing_data >0])

# Check for duplicates
print("\nDuplicate Rows: ", df.duplicated().sum())
```

Data type Conversion 

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%Y-%m-%d')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%Y-%m-%d')
```

UNIVARIATE ANALYSIS

```python
df.info()
```

```python
#  univariate analysis is the simplest form of statisical analysis. As the name suggests- "uni"
#means "one,"and "variate" refers to a single variable. univariate analysis is the examination of the distribution of a single variable, without considering any relationsships with other variables. The main purpose of univariate and analysis is to describe the data and find patterns that exist with in the variable.

print("\n === UNIVERIATE ANALYSIS === \n")

categorical_cols = ['Customer Segment', 'Category', 'Region', 'Payment Mode', 'Delivery Status', 'Auto Reorder']

for col in categorical_cols:
    
    print(f'\n--- {col} ---\n')
    print(df[col].value_counts())


    #visualization 
    plt.figure(figsize=(6, 6))
    df[col].value_counts().plot(kind='bar',color='skyblue')
    plt.title(f'distribution of {col}')
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
    
```

Numerical Variable Analysis

```python
print("\n=== NUMERICAL VARIABLES ANALYSIS ===\n")

numerical_cols = ["Quantity", "Unit Price", 'Discount (%)', 'Sales Amount', 'Profit', 'Stock Left', 'Reorder Quantity']

for col in numerical_cols:
    print(f"\n--- {col} ---\n")
    print(f"Mean:{df[col].mean():.2f}")
    print(f"Median:{df[col].median():.2f}")
    print(f"Std Dev:{df[col].std():.2f}")
    print(f"Min:{df[col].min():.2f}")
    print(f"Max:{df[col].max():.2f}")

    # Visualization 
    plt.figure(figsize=(8, 4))

    plt.subplot(1, 2, 1)
    df[col].hist(bins=30, color='lightgreen', edgecolor='black' , alpha=0.7)
    plt.title(f'Histogram of {col}')

    plt.subplot(1, 2, 2)
    df.boxplot(column = col)
    plt.title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()
```

BIVARIATE ANALYSIS

```python
# bivariate analysis is the statistical analysis of two variables at the same time. The core purpose is to investigate the relationship between them.


print("\n=== BIVARIATE ANALYSIS ===\n")
print("\n=== SALES & PROFIT BY CATEGOTRY ===\n")

categorical_performance = df.groupby('Category').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'
}).rename(columns={'Order ID': 'Total Orders'})


categorical_performance['Profit Margin'] = (categorical_performance['Profit'] / categorical_performance['Sales Amount']) * 100
print(categorical_performance,'\n\n')

# Visualization
fig ,axes = plt.subplots(2, 2, figsize=(15, 12))

#sales by category
categorical_performance['Sales Amount'].plot(kind='bar', ax=axes[0, 0], color='blue', alpha=0.7)
axes[0, 0].set_title('Total Sales by Category')
axes[0, 0].set_xlabel('Category')
axes[0, 0].set_ylabel('Total Sales')
axes[0, 0].tick_params(axis='x', rotation=30)

# profit by category
categorical_performance['Profit'].plot(kind='bar', ax=axes[0, 1], color='green', alpha=0.7)
axes[0, 1].set_title('Total Profit by Category')
axes[0, 1].set_xlabel('Category')
axes[0, 1].set_ylabel('Total Profit')
axes[0, 1].tick_params(axis='x', rotation=30)

# total orders by category
categorical_performance['Total Orders'].plot(kind='bar', ax=axes[1, 0], color='lightgreen', alpha=0.7)
axes[1, 0].set_title('Total Orders by Category')
axes[1, 0].set_xlabel('Category')
axes[1, 0].set_ylabel('Total Orders')   
axes[1, 0].tick_params(axis='x', rotation=30)

# Profit margin by category
categorical_performance['Profit Margin'].plot(kind='bar', ax=axes[1, 1], color='orange', alpha=0.7)
axes[1, 1].set_title('Profit Margin by Category (%)')
axes[1, 1].set_xlabel('Category')
axes[1, 1].set_ylabel('Profit Margin (%)')
axes[1, 1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

Regional Performance

```python
# bivariate analysis is the statistical analysis of two variables at the same time. The core purpose is to investigate the relationship between them.


print("\n=== REGIONAL PERFORMANCE===\n")


regional_performance = df.groupby('Region').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'
}).rename(columns={'Order ID': 'Total Count'})


regional_performance['Profit Margin'] = (regional_performance['Profit'] / regional_performance['Sales Amount']) * 100
print(regional_performance,'\n\n')

# Visualization
fig ,axes = plt.subplots(2, 2, figsize=(15, 12))

#sales by regional
plt.subplot(2,2,1)
regional_performance['Sales Amount'].plot(kind='pie', autopct ="%1.1f")
plt.title('Sales Distribution by Region')


#Profitby regional
plt.subplot(2,2,2)
regional_performance['Profit'].plot(kind='pie', autopct ="%1.1f")
plt.title('Profit Distribution by Region')

#Order Count by regional
plt.subplot(2,2,3)
sns.barplot(x=regional_performance.index ,y='Total Count' ,data =regional_performance , color='lightgreen')
plt.title('Order Count by Region')
plt.xticks(rotation=30)

#Profit Margin by regional
plt.subplot(2,2,4)
sns.barplot(x=regional_performance.index ,y='Profit Margin' ,data =regional_performance , color='orange')
plt.title('Profit Margin by Region')
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

TIME SERIES ANALYSIS

```python
print("\n=== TIME SERIES ANALYSIS ===\n")

#extract time components 
df['Order Month'] = df['Order Date'].dt.to_period('M')
df['Order Year'] = df['Order Date'].dt.year

# Monthly trends
monthly_trends = df.groupby('Order Month').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'
}).rename(columns={'Order ID': 'Total Orders'})

print('Monthly Trends:')
print(monthly_trends.tail(10),'\n\n')

#visualization
plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 1)
monthly_trends['Sales Amount'].plot(kind='line', marker='o', color='purple')
plt.title('Monthly Sales Trends')
plt.ylabel('Sales Amount')
plt.grid(True)

plt.subplot(3, 1, 2)
monthly_trends['Profit'].plot(kind='line', marker='o', color='green')
plt.title('Monthly Profit Trends')
plt.ylabel('Profit')
plt.grid(True)

plt.subplot(3, 1, 3)
monthly_trends['Total Orders'].plot(kind='line', marker='o', color='pink')
plt.title('Monthly Order Trends')   
plt.ylabel('Total Orders')
plt.grid(True)

plt.tight_layout()
plt.show()
```

ADVANCED ANALYSIS

```python
print("\n==== CORRELATION ANALYSIS ====\n")

#select numerical columns for correlation
numerical_df = df[['Quantity', 'Unit Price', 'Discount (%)', 'Sales Amount','Cost Price' , 'Profit']]

# correlation matrix
correlation_matrix = numerical_df.corr()
print("Correlation Matrix:\n", correlation_matrix,'\n\n')

# Visualization of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5 , square=True)
plt.title("Correlation Matrix Heatmap")
plt.tight_layout()
plt.show()  
```

Customer Segment Analysis

```python
print("\n=== CUSTOMER SEGMENT ANALYSIS ===\n")

segment_analysis = df.groupby('Customer Segment').agg({
    'Sales Amount': ['sum', 'mean'],
    'Profit': ['sum', 'mean'],
    'Order ID': 'count',
    'Discount (%)': 'mean'
}).round(2)

segment_analysis.columns = [
    'Total Sales',
    'Average Sales',
    'Total Profit',
    'Average Profit',
    'Total Orders',
    'Average Discount'
]

print(segment_analysis, "\n\n")


# VISUALIZATION
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Total Sales
segment_analysis['Total Sales'].plot(kind='bar', ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Total Sales by Customer Segment')
axes[0, 0].set_xlabel('Customer Segment')
axes[0, 0].set_ylabel('Total Sales')
axes[0, 0].tick_params(axis='x', rotation=30)

# Average Sales
segment_analysis['Average Sales'].plot(kind='bar', ax=axes[0, 1], color='skyblue')
axes[0, 1].set_title('Average Sales by Customer Segment')
axes[0, 1].set_xlabel('Customer Segment')
axes[0, 1].set_ylabel('Average Sales')
axes[0, 1].tick_params(axis='x', rotation=30)

# Total Profit
segment_analysis['Total Profit'].plot(kind='bar', ax=axes[0, 2], color='green')
axes[0, 2].set_title('Total Profit by Customer Segment')
axes[0, 2].set_xlabel('Customer Segment')
axes[0, 2].set_ylabel('Total Profit')
axes[0, 2].tick_params(axis='x', rotation=30)

# Average Profit
segment_analysis['Average Profit'].plot(kind='bar', ax=axes[1, 0], color='purple')
axes[1, 0].set_title('Average Profit by Customer Segment')
axes[1, 0].set_xlabel('Customer Segment')
axes[1, 0].set_ylabel('Average Profit')
axes[1, 0].tick_params(axis='x', rotation=30)

# Total Orders
segment_analysis['Total Orders'].plot(kind='bar', ax=axes[1, 1], color='lightgreen')
axes[1, 1].set_title('Total Orders by Customer Segment')
axes[1, 1].set_xlabel('Customer Segment')
axes[1, 1].set_ylabel('Total Orders')
axes[1, 1].tick_params(axis='x', rotation=30)

# Average Discount
segment_analysis['Average Discount'].plot(kind='bar', ax=axes[1, 2], color='orange')
axes[1, 2].set_title('Average Discount by Customer Segment')
axes[1, 2].set_xlabel('Customer Segment')
axes[1, 2].set_ylabel('Average Discount (%)')
axes[1, 2].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

KEY INSIGHTS & SUMMARY

```python
print("\n=== KEY INSIGHTS & SUMMARY ===\n")

#Overall matrics
total_sales = df['Sales Amount'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
avg_profit_margin = (total_profit / total_sales) * 100

print(f"OVERALL METRICS:\n")
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Total Orders: {total_orders}")
print(f"Average Profit Margin: {avg_profit_margin:.2f}%\n\n")

# top performance
top_categories = df.groupby('Category')['Sales Amount'].sum().sort_values(ascending=False)
print("Top Performing Categories : \n")
for i , ( category, profit) in enumerate(top_categories.items(), start=1):
    print(f"{i}. {category}: ${profit:,.2f}")


# Regional performance
top_regions =df.groupby('Region')['Profit'].sum().idxmax()
print(f"\nTop Performing Regions : {top_regions}" )

# Delivery performance
devilery_stats = df['Delivery Status'].value_counts(normalize=True) * 100
print("\nDelivery Status:\n")
for status, percentage in devilery_stats.items():
    print(f"{status}: {percentage:.2f}%")


# customer segment performance
best_segment =df.groupby('Customer Segment')['Profit'].sum().idxmax()
print(f"\nBest Performing Customer Segment: {best_segment}")
```

EXPORT RESULT

```python
analysis_results = {
    "total_sales": total_sales,
    "total_profit": total_profit,
    "total_orders": total_orders,
    "avg_profit_margin": avg_profit_margin,
    "top_categories": top_categories.index[0],
    "top_regions": top_regions,
    "delivery_stats": devilery_stats.to_dict(),
    "best_segment": best_segment
}

#convert to dataframe and save
result_df = pd.DataFrame([analysis_results])
result_df.to_csv('superstore_analysis_results.csv', index=False)

```
